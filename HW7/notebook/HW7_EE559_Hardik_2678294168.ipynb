```python
import numpy as np
import pandas as pd
import warnings
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from metric_learn import MLKR
from sklearn import metrics
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
from sklearn.kernel_ridge import KernelRidge

warnings.filterwarnings('ignore')
```

```python
col=['subject#','age','sex','test_time','motor_UPDRS','total_UPDRS','Jitter(%)','Jitter(Abs)','Jitter:RAP','Jitter:PPQ5','Jitter:DDP','Shimmer','Shimmer(dB)','Shimmer:APQ3','Shimmer:APQ5','Shimmer:APQ11','Shimmer:DDA','NHR','HNR','RPDE','DFA','PPE']
df=pd.read_csv("../data/parkinsons_updrs.data")
df
```

```python
#split test-train data 
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,6:], df.iloc[:,4:6], test_size=0.30, random_state=42)
print("Training data is of length {} ".format(len(X_train)))
print("Testing data is of length {} ".format(len(X_test)))
print(X_train.head(5))
print(y_train.head(5))
```

```python
#standardizing data
sc = StandardScaler()
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.fit_transform(X_test)
y_train_std=sc.fit_transform(y_train)
y_test_std=sc.fit_transform(y_test)
y_train_std=pd.DataFrame(y_train_std)
y_test_std=pd.DataFrame(y_test_std)
X_train_std=pd.DataFrame(X_train_std)
X_test_std=pd.DataFrame(X_test_std)

print(X_train_std)
print(y_train_std)
```

```python
#seperating the labels of test and train data
y_train_motor_std=y_train_std.iloc[:,0]
y_train_total_std=y_train_std.iloc[:,1]
y_test_motor_std=y_test_std.iloc[:,0]
y_test_total_std=y_test_std.iloc[:,1]

y_train_motor_std=pd.DataFrame(y_train_motor_std)
y_train_total_std=pd.DataFrame(y_train_total_std)
y_test_motor_std=pd.DataFrame(y_test_motor_std)
y_test_total_std=pd.DataFrame(y_test_total_std)

print(y_train_motor_std)

```

Q4_b: Use metric learning with Gaussian kernels to estimate each of the outputs motor UPDRS and total UPDRS from the features.


```python
print(" Train and test scores for Gaussian Kernels : motor UPDRS")
for k in np.array([5,10,15,16]):
    mlkr=MLKR(n_components=k,init='auto', max_iter=10, verbose=False)
    mlkr.fit(X_train_std,y_train_motor_std)
    X_train_transform=mlkr.transform(X_train_std)
    clf_kernel = KernelRidge(alpha=1.0,kernel='rbf')
    clf_kernel.fit(X_train_transform, y_train_motor_std)
    X_test_transform=mlkr.transform(X_test_std)
    train_score_kernel=clf_kernel.score(X_train_transform,y_train_motor_std)
    test_score_kernel=clf_kernel.score(X_test_transform,y_test_motor_std)
    print("No. of features : {}".format(k))
    print("Train score = {} ".format(train_score_kernel))
    print("Test score = {} ".format(test_score_kernel))
    print(" ")
```

```python
print(" Train and test scores for Gaussian Kernels : total UPDRS")
for k in np.array([5,10,15,16]):
    mlkr=MLKR(n_components=k,init='auto', max_iter=10, verbose=False)
    mlkr.fit(X_train_std,y_train_total_std)
    X_train_transform=mlkr.transform(X_train_std)
    clf_kernel = KernelRidge(alpha=1.0,kernel='rbf')
    clf_kernel.fit(X_train_transform, y_train_total_std)
    X_test_transform=mlkr.transform(X_test_std)
    train_score_kernel=clf_kernel.score(X_train_transform,y_train_total_std)
    test_score_kernel=clf_kernel.score(X_test_transform,y_test_total_std)
    print("No. of features : {}".format(k))
    print("Train score = {} ".format(train_score_kernel))
    print("Test score = {} ".format(test_score_kernel))
    print(" ")
```

It can be observed that when using 10 features, we had the best training as well as testing scores (r2_score) for both the outputs. 
Also, when compared between the 2 output scores, the training and test scores were better for 'total UPDRS'

Q4_c:Use sklearn's neural network implementation to train a neural network with two
outputs that predicts motor UPDRS and total UPDRS. Use a single layer

```python
clf_Neural={}
def neural_paramSearch(model,xtrain,ytrain,xtest,ytest,parameters):
    classifier = GridSearchCV(estimator=model,**parameters)
    classifier.fit(xtrain,ytrain)
    average=classifier.cv_results_['mean_test_score']
    deviation=classifier.cv_results_['std_test_score']
    print("The grid scores of cross validation for coefficients are : ")
    for mean,std,para in zip(average, deviation,classifier.cv_results_['params']):
        print (' %0.3f (+/- %0.3f) : Coeff %r' %(mean,std,para))
    print(" .")
    print("The optimal coefficients are : {}".format(classifier.best_params_))
    predicty_test=classifier.predict(xtest)
    print(" ")
    return classifier,predicty_test


coeff = {'hidden_layer_sizes' : np.linspace(50,300,6).astype(int),'alpha':np.linspace(0.0001,0.01,10)}
cv_neural = KFold(5, random_state=4112, shuffle=True)
neural_parameters = {
    'param_grid' : coeff, 
    'cv' : cv_neural,
    'scoring' : 'r2', 
    'verbose' : 1
}

```

```python
clf_mode=MLPRegressor(activation='relu',solver='lbfgs')
clf_neural_network=neural_paramSearch(clf_mode,X_train_std.values,y_train_std.values,X_test_std.values,y_test_std.values,neural_parameters)
```

```python
#printinng test and train scores for neural network with best parameters found
print('layer_size=300, alpha:0.0001 : found from cross validation')
print('activation: relu :- rectified linear works best for regression (educated guess)')
print('solver: lbfgs :-fits well and quick for low size data set as is our case')
best_parameters=clf_neural_network[0].best_params_
l2_para=best_parameters['alpha']
hidden_layer_size=best_parameters['hidden_layer_sizes']
best_neural_network=MLPRegressor(activation='relu',solver='lbfgs',hidden_layer_sizes=hidden_layer_size,alpha=l2_para)
best_neural_network.fit(X_train_std.values,y_train_std.values)
train_score=best_neural_network.score(X_train_std.values, y_train_std.values)
test_score=best_neural_network.score(X_test_std.values, y_test_std.values)
print("The train and test scores for best parameters are : {} and {} respectively".format(train_score,test_score))
```

```python
best_neural_earlystop=MLPRegressor(activation='relu',solver='lbfgs',hidden_layer_sizes=hidden_layer_size,alpha=l2_para,early_stopping=True)
best_neural_earlystop.fit(X_train_std.values,y_train_std.values)
train_score=best_neural_earlystop.score(X_train_std.values, y_train_std.values)
test_score=best_neural_earlystop.score(X_test_std.values, y_test_std.values)
print("The train and test scores for best parameters with early-stopping are : {} and {} respectively".format(train_score,test_score))
```
